```python
# Osprey — GPU Training Dashboard

import cudf
import matplotlib.pyplot as plt
from data.unified_loader import load_all
from features.gpu_vector_builder import build_gpu_vector
from cuml.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score

df = load_all()
gdf = cudf.DataFrame(df)

X = build_gpu_vector(gdf)
y = gdf["label"]
```

```python
model = RandomForestClassifier(n_estimators=500)
model.fit(X, y)

preds = model.predict_proba(X)[:, 1]
auc = roc_auc_score(y.to_pandas(), preds.get())
auc
```

```python
plt.hist(preds.get(), bins=50)
plt.title("GPU Model Score Distribution")
plt.xlabel("Score")
plt.ylabel("Count")
```
